```python
import osmnx as ox

# Requête sur Paris
place = "Paris, France"

# Récupérer les entités 'waterway' (cours d’eau) dans Paris
waterways = ox.features_from_place(place, tags={"waterway": "river"})

# Filtrer la Seine
seine = waterways[waterways["name"] == "La Seine"]

# Afficher les coordonnées GPS de la géométrie
print(seine.geometry.iloc[0])
```

```python
import matplotlib.pyplot as plt

# 4️⃣ Tracer la carte de Paris avec la Seine
fig, ax = plt.subplots(figsize=(8, 8))

# Afficher le fond de carte (routes principales par exemple)
graph = ox.graph_from_place(place, network_type="drive")
ox.plot_graph(graph, ax=ax, show=False, close=False, node_size=0, edge_color="lightgray")

# Tracer la Seine en bleu
seine.plot(ax=ax, color="blue", linewidth=2)

# Titre et affichage
plt.title("La Seine à Paris (coordonnées GPS issues d'OpenStreetMap)")
plt.show()
```
